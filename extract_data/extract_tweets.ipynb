# main functions

```python
def extract_ids_from(csv_path, id_index):
  ids = []
  # opening the csv file:
  with open(csv_path, mode='r') as file:
      csvFile = csv.reader(file)
      for lines in csvFile:
          if(lines[id_index] != ''):
            ids.append(lines[id_index])
  del ids[0] # bc it is the header
  return ids
```

```python
def convert_ids_to_str(ids):
  id_lists = [ids[x:x+100] for x in range(0, len(ids), 100)]
  # convert each id_list to string:
  id_lists_str = []
  for id_list in id_lists:  
    id_list_str = ""
    for i in id_list:
        id_list_str += i 
        id_list_str += ','
    id_list_str = id_list_str.rstrip(id_list_str[-1])
    id_lists_str.append(id_list_str)
  return id_lists_str
```

```python
def get_all_tweets(id_lists_str):
  all_tweets = {'id':[] , 'text':[]}
  for id_list in id_lists_str:  
    # get tweets info:
    tweets = client.get_tweets(id_list)
    tweets_data = {'id': [tweet.id for tweet in tweets.data if tweet.id], 'text': [tweet.text for tweet in tweets.data if tweet.text]}  
    all_tweets['id'].extend(tweets_data['id'])
    all_tweets['text'].extend(tweets_data['text'])
  return all_tweets
```

# Extract Users

## Functions:

```python
def get_all_users(id_lists_str):
  users_info_sarcastic = {'user_id':[], 'name':[], 'username':[]}
  for id_list in id_lists_str:  
    tweets = client.get_tweets(id_list, expansions ='author_id')
    users_data = {'user_id': [user.id  if user.id else None for user in tweets.includes['users']],
                    'name': [user.name  if user.name else None for user in tweets.includes['users']],
                    'username': [user.username if user.username else None for user in tweets.includes['users']]}  
    users_info_sarcastic['user_id'].extend(users_data['user_id'])
    users_info_sarcastic['name'].extend(users_data['name'])
    users_info_sarcastic['username'].extend(users_data['username'])
  return users_info_sarcastic
```

```python
def get_users_one_by_one(id_list):
  users_info_sarcastic = {'tweet_id':[] , 'user_id':[], 'name':[], 'username':[]}
  ids = id_list.split(',')
  for id in ids: 
    tweet = client.get_tweet(id, expansions ='author_id')
    if tweet.data and tweet.includes['users']:
      users_info_sarcastic['tweet_id'].extend([tweet.data.id])
      users_info_sarcastic['user_id'].extend([tweet.includes['users'][0]['id']])
      users_info_sarcastic['name'].extend([tweet.includes['users'][0]['name']])
      users_info_sarcastic['username'].extend([tweet.includes['users'][0]['username']])
  return users_info_sarcastic
```

```python
def get_all_users_with_tweets(id_lists_str):
  users_info_sarcastic = {'tweet_id':[],'user_id':[], 'name':[], 'username':[]}
  failed_id_lists = []
  for id_list in id_lists_str:  
    tweets = client.get_tweets(id_list, expansions ='author_id')
    users_data = {'tweet_id': [tweet.id for tweet in tweets.data],
                    'user_id': [user.id  if user.id else None for user in tweets.includes['users']],
                    'name': [user.name  if user.name else None for user in tweets.includes['users']],
                    'username': [user.username if user.username else None for user in tweets.includes['users']]}  
    if len(tweets.data) == len(tweets.includes['users']):
      users_info_sarcastic['tweet_id'].extend(users_data['tweet_id'])
      users_info_sarcastic['user_id'].extend(users_data['user_id'])
      users_info_sarcastic['name'].extend(users_data['name'])
      users_info_sarcastic['username'].extend(users_data['username'])
    else:
      failed_id_lists.append(id_list)
  return users_info_sarcastic, failed_id_lists
```
